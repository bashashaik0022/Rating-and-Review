```python
#Importing Libary

import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
#Driver

driver = webdriver.Chrome(r"C:\Users\shaik mahaboob basha\Downloads\chromedriver.exe")
#driver.get('https://www.flipkart.com/')
```

# Amazon Web Scraping

```python
#Scaraping Data from Amazon

prod = ['Laptops']

#Getting driver
driver.get('https://www.amazon.in/')

for pr in prod:
    #Searching Laptops in the search bar and clicking the search button
    search_bar=driver.find_element_by_id('twotabsearchtextbox')
    search_bar.send_keys(pr)
    
    driver.find_element_by_id('nav-search-submit-button').click()

#Creating empty lists
Product_URL=[]
Ratings=[]
Review=[]
Product = [] 
#Getting URLs of the product
for i in range(1,4):
    URL = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-medium']/div[2]/div[2]/div/div/h2/a")
    for i in URL:
        Product_URL.append(i.get_attribute('href'))
    try:
        next_btn=driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    except NoSuchElementException:
        pass
    
    #time.sleep(1)
    
for i in Product_URL:
    driver.get(i)
    
    #Clicking the rating
    try:
        driver.find_element_by_xpath("//a[1][@id='acrCustomerReviewLink']").click()
    except NoSuchElementException:
        print("No rating")
        pass
    
    #Clicking to see all reviews
    try:
        driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
    except NoSuchElementException:
        pass
    
    #time.sleep(1)
    
    #Scrapping the details
    Start_page=1
    End_page=50
    for page in range(Start_page,End_page+1):
        try:
            Reviews=driver.find_elements_by_xpath("//div[@class='a-row a-spacing-small review-data']/span/span")
            for r in Reviews:
                Review.append(r.text.replace('\n',''))
        except NoSuchElementException:
            Review.append("Not Available")
        try:
            Rating=driver.find_elements_by_xpath("//div[@class='a-section celwidget']/div[2]/a[1]")
            for i in Rating:
                rating=i.get_attribute('title')
                Ratings.append(rating[:3])
        except NoSuchElementException:
            Ratings.append("Not available")      
        
        try:
            Product.append(pr)
        except NoSuchElementException   as e:
            Product.append(pr)#appending the No review if no rating is there
        
        
        #time.sleep(1)
        
        #Looping for going to next page automatically
        try:
            next_page=driver.find_element_by_xpath("//div[@id='cm_cr-pagination_bar']/ul/li[2]/a")
            if next_page.text=='Next Page':
                next_page.click()
                time.sleep(2)
        except NoSuchElementException:
            pass

```

```python
#Checking the length of data extracted
print(len(Review),len(Ratings),len(Product))
```

```python
#Saving in dataframe
Laptops=pd.DataFrame({'Product_Review':Review[:950],'Ratings':Ratings[:950],'Product_Name':Product[:950]})
Laptops
```

```python
rating=[]
full_review=[]
product=[]
for k in urls:
    driver.get(k)
    time.sleep(3)
    
    try:
        driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click() # clcik on see all reviews
        time.sleep(2)
        # fetching first pages reviews
        for a in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
            full_review.append(a.text)
        for b in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
            rating.append(b.text)
                   
            
        time.sleep(1)
        #clciking on second page for reviews
        driver.find_element_by_xpath("//a[@class='_1LKTO3']/span").click()
        time.sleep(2)
        # running loop till the end of pages of all reviews
        while True:
            time.sleep(2)
            
            for i in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
                full_review.append(i.text)
            for j in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):#"//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
                rating.append(i.text)
                               
            try:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]/span").click()
                time.sleep(1)
            except:
                break
    except:
        print("no reviews")
```
